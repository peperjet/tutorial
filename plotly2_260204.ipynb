```python
import matplotlib.pyplot as plt

# 1. 데이터 준비
x = [1, 2, 3, 4, 5]
y = [10, 25, 30, 45, 50]

# 2. 그래프 그리기 (선 그래프)
plt.figure(figsize=(8, 5)) # 그래프 크기 설정
plt.plot(x, y, marker='o', linestyle='--', color='b', label='성장 지표')

# 3. 꾸미기
plt.title('Sample Graph on Colab')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)

# 4. ★핵심★ 출력 명령
plt.show()
```

```python
import plotly.express as px
import pandas as pd

# 샘플 데이터 (실제 데이터 df.head(100) 등으로 대체 가능)
df = pd.DataFrame({
    "date": ["2023-01-01", "2023-01-02", "2023-01-03", "2023-01-04"],
    "confirmed": [100, 150, 130, 200],
    "country": ["Korea", "Korea", "Korea", "Korea"]
})

fig = px.line(df, x="date", y="confirmed", title="코비드19 확진자 추이", markers=True)
fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

# 국가별 데이터 추가
fig.add_trace(go.Scatter(x=[1, 2, 3], y=[10, 20, 30], name="Korea"))
fig.add_trace(go.Scatter(x=[1, 2, 3], y=[50, 40, 60], name="USA", visible=False))

# 드롭다운 메뉴 설정
fig.update_layout(
    updatemenus=[{
        "buttons": [
            {"label": "Korea 보기", "method": "update", "args": [{"visible": [True, False]}]},
            {"label": "USA 보기", "method": "update", "args": [{"visible": [False, True]}]},
        ],
        "direction": "down",
        "showactive": True,
    }],
    title="국가 선택 드롭다운"
)
fig.show()
```
