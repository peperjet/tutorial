```python
# 표 만들기

import pandas as pd

data = {
    "name" : ["A", "B", "C"],
    "score" : [80, 90, 70]
}


df = pd.DataFrame(data)
print(df)


# 열 하나 선택
df["score"]


# 평균 구하기
df["score"].mean()


#조건 필터
df[df["score"] >= 80]
```

```python
# 매출 데이터 정리하기

import pandas as pd
data = {
    "date" : ["2026-01-01", "2026-01-01","2026-01-02", "2026-01-02"],
    "product" : ["A", "B", "A", "B"],
    "sales" : [10000, 15000, 12000, 13000]
}

df = pd.DataFrame(data)


# 1. 날짜별 총 매출
daily_sales = df.groupby("date")["sales"].sum()
print(daily_sales)

# 2. 상품 A만 필터
product_a = df[df["product"] == "A"]
print(product_a)

# 3. 평균 매출
avg_sales = df["sales"].mean()
print(avg_sales)
```

```python
# 날짜별 매출 요약

import pandas as pd

df = pd.DataFrame({
    "date" : ["2026-01-01", "2026-01-01","2026-01-02", "2026-01-02"],
    "product" : ["A", "B", "A", "B"],
    "sales" : [10000, 15000, 12000, 13000]
})

# 날짜
daily = df.groupby("date", as_index=False)["sales"].sum()
print(daily)

# 안티그래비티 용으로 쓰기 좋게
daily.to_dict(orient="records")

# 매출 높은 날짜순
daily.sort_values("sales", ascending=False)

#상품별 요약도
by_product = df.groupby("product", as_index=False)["sales"].sum()
print(by_product)


# 막대 그래프
import matplotlib.pyplot as plt

# daily가 이미 만들어져 있다고 가정
plt.figure(figsize=(6,4))
plt.bar(daily["date"], daily["sales"])
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Daily Sales")
plt.show()


# 색 변경
plt.figure(figsize=(6,4))
plt.bar(
  daily["date"],
  daily["sales"],
  color="purple" #색변경
)
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Daily Sales")
plt.show()
```
