```python
# Plotly 그래프 생성
# iris 데이터 산점도 (기본)

import plotly.express as px

df = px.data.iris()

fig = px.scatter(
    df,
    x="sepal_width",
    y="sepal_length",
    color="species",
    title="Iris 산점도"
)
fig.show()



# 막대그래프
import plotly.express as px

df = px.data.iris()
fig = px.bar(df, x="species", y="petal_length", title="종별 petal_length")
fig.show()
```

```python
#Plotly trace 업데이트 (express - trace 스타일 바꾸기)

import plotly.express as px

df = px.data.iris()
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species")

fig.update_traces(marker=dict(size=10, opacity=0.5)) #점크기, 투명도
fig.show()


# 특정 trace 바꾸기
fig.update_traces(marker=dict(size=14), selector=dict(type="scatter"))
```

```python
# Plotly xaxes, yaxed (축 설정)

import plotly.express as px

df = px.data.iris()
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species")

fig.update_xaxes(title="꽃받침 너비", showgrid=True)
fig.update_yaxes(title="꽃받침 길이", showgrid=True)

fig.update_layout(title="축 설정 예시", width=800, height=500)
fig.show()
```

```python
# 캐글을 이용한 선 그래프 (주식가격 추이) all_stock_5yr.csv

# csv 읽기 + 한 회사만 읽기 + close 선 그래프

import pandas as pd
import plotly.express as px

df = pd.read_csv("all_stocks_5yr.csv")

one = df[df["Name"] == "AAPL"].copy()
one["date"] = pd.to_datetime(one["date"])

fig = px.line(one, x="date", y="close", title="AAPL 종가 추이")
fig.show()
```
