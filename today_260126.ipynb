```python
import matplotlib.pyplot as plt

# 1. 데이터 준비 (영문으로)
names = ['A', 'B', 'C']
score = [80, 100, 70]

# 2. 막대 그래프 그리기
plt.bar(names, score, color='orange')

# 3. 그래프 제목 정하기
plt.title("Simple Bar Chart")

# 4. 화면에 보여줘!
plt.show()
```

```python
# 선 그래프 (Stock Price Trend)

# 1. 데이터 준비 (날짜와 가격)
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
price = [150, 160, 155, 175, 180]

# 2. 선 그래프 그리기 (plot은 점들을 선으로 이어줍니다)
plt.plot(days, price, marker='o', color='green', linestyle='--')

# 3. 그래프 꾸미기
plt.title("My stock price trend") #제목
plt.xlabel("Days") # x축이름
plt.ylabel("Price") #y축이름


# 4. 출력
plt.show()
```

```python
# 워드클라우드로 텍스트 데이터 시각화하기

from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 1. 시각화하고 싶은 텍스트(단어들)
text = "python data analysis visual graph plot python pandas seaborn python"

# 2. 워드클라우드 생성 (배경색과 크기 설정)
wc = WordCloud(width=400, height=400, background_color='white').generate(text)

# 3. 그래프로출력 (워드클라우드는 이미지라 imshow를 써요)
plt.imshow(wc)
plt.axis('off') #테두리 숫자와 눈금 끄기
plt.show()
```

```python
# 마지막. 미니 코로나 대시보드

import matplotlib.pyplot as plt

# 1. 데이터 준비 (날짜별 확진자 수 가상 데이터)
dates = ['01-20', '01-21', '01-22', '01-23', '01-24']
confirmed = [150, 200, 180, 300, 250] #확진자 수
recovered = [50, 80, 100, 120, 150] #완치자 수

# 2. 대시보드 화면 분할 (1행 2열로 구성해볼게요)
plt.figure(figsize=(12, 5)) #전체 화면 크기 설정

# --- 왼쪽 : 선 그래프 (확진자 추이) ---
plt.subplot(1, 2, 1) # (1행 2열중 첫번째)
plt.plot(dates, confirmed, marker='o', color='red', label='Confirmed')
plt.title("COVID-19 Confirmed Cases")
plt.legend()

# 3. 전체 출력
plt.tight_layout() #그래프끼리 겹치지 않게 자동 조정
plt.show()

```

```python
# 판다스 데이터프레임 생성하기

import pandas as pd

# 데이터 준비 (이름, 국어, 영어 점수)
data = {
    'name' : ['Alice', 'Bob', 'Charlie', 'Gemini'],
    'Korean' : [80, 90, 70, 100],
    'English' : [85, 95, 75, 90]
}


# 데이터 프레임 생성
df = pd.DataFrame(data)

# 결과 확인
print(df)
```

```python
# 판다스 사칙 연산 수행하기 - 모든 행에 대해 한 번에 계산이 가능하다.

# 새로운 컬럼 'Total'(합계)만들기
df['Total'] = df['Korean'] + df['English']

# 새로운 컬럼 'Average'(평균)만들기
df['Average'] = df['Total']/2

print(df)

# 판다스 기타 연산 메서드 활용하기
print("--- 통계 요약---")
print(f"국어 최고점: {df['Korean'].max()}")
print(f"영어 평균점: {df['English'].mean()}")
print(f"전체 점수 합계: {df['Total'].sum()}")

# 데이터의 전반적인 통계를 한눈에!
print(df.describe())
```

```python
# 이어서 기타 연산 메서드

# 1. 특정 조건에 맞는 데이터 세기 (예: 특정 점수대의 학생 수)
print("--- 개수 세기 ---")
print(df['Korean'].value_counts())

# 2. 데이터 정렬하기 (점수 높은 순으로!)
print("\n--- 점수순 정렬---")
sorted_df = df.sort_values(by='Total', ascending=False)
print(sorted_df)

# 3. 데이터 상단/하단만 보기 (데이터가 많을 때 필수!)
print("\n--- 상단 2개만 보기---")
print(df.head(2))
```
